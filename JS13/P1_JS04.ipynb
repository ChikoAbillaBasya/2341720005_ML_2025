Praktikum 1 ini akan membuat JST sederhana (2 layer) dengan forward pass dan backpropagation manual.

Backpropagation adalah sebuah algoritma untuk melatih jaringan saraf tiruan dengan cara mengoreksi kesalahan. Algoritma ini bekerja dengan cara menghitung selisih antara keluaran yang dihasilkan jaringan dan keluaran yang seharusnya (kesalahan), lalu memperbarui bobot dan bias jaringan secara berulang dari keluaran ke masukan untuk meminimalkan kesalahan tersebut. Cara kerja backpropagation
* **Perambatan maju (*forward pass*)**: Data masukan diproses melalui jaringan dari lapisan masukan ke lapisan keluaran untuk menghasilkan prediksi awal.
* **Hitung kesalahan**: Selisih antara keluaran prediksi dan keluaran target dihitung menggunakan [fungsi kerugian](https://www.google.com/search?q=fungsi+kerugian&oq=backpropagation+adalah&gs_lcrp=EgZjaHJvbWUyBggAEEUYOTIGCAEQLhhA0gEINzYwNWowajGoAgiwAgE&sourceid=chrome&ie=UTF-8&ved=2ahUKEwjLoei9vvOQAxVC2TgGHcLNNxUQgK4QegYIAQgAEAo).
* **Perambatan mundur (*backward pass*)**: Kesalahan disebarkan kembali ke belakang melalui jaringan dari lapisan keluaran ke lapisan masukan untuk menghitung gradien atau turunan parsial dari fungsi kerugian terhadap bobot dan bias.
* **Perbarui bobot**: Bobot dan bias disesuaikan menggunakan algoritma penurunan gradien untuk mengurangi kesalahan pada iterasi berikutnya.

**Langkah:**

1. Buat dataset sederhana (XOR).
2. Inisialisasi bobot dan bias.
3. Implementasikan forward pass.
4. Hitung error dan lakukan backpropagation.
5. Update bobot menggunakan gradient descent.

```python
import numpy as np

# Dataset XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Parameter
input_size = 2
hidden_size = 2
output_size = 1
lr = 0.1

# Inisialisasi bobot
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

# Fungsi aktivasi
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Training
for epoch in range(10000):
    # Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Hitung error
    error = y - a2

    # Backpropagation
    d_a2 = error * sigmoid_derivative(a2)
    d_W2 = np.dot(a1.T, d_a2)
    d_b2 = np.sum(d_a2, axis=0, keepdims=True)

    d_a1 = np.dot(d_a2, W2.T) * sigmoid_derivative(a1)
    d_W1 = np.dot(X.T, d_a1)
    d_b1 = np.sum(d_a1, axis=0, keepdims=True)

    # Update bobot
    W1 += lr * d_W1
    b1 += lr * d_b1
    W2 += lr * d_W2
    b2 += lr * d_b2

    if epoch % 1000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss}")

# Output akhir
print("Prediksi:")
print(a2)
```

* Bandingkan hasil loss dengan konfigurasi awal.

```python
import numpy as np

# Dataset XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# -------------------------------------------
# Fungsi untuk training dan mengembalikan loss
# -------------------------------------------
def train_model(hidden_size, lr=0.1, epoch_total=10000):
    input_size = 2
    output_size = 1

    # Inisialisasi bobot
    W1 = np.random.randn(input_size, hidden_size)
    b1 = np.zeros((1, hidden_size))
    W2 = np.random.randn(hidden_size, output_size)
    b2 = np.zeros((1, output_size))

    for epoch in range(epoch_total):

        # Forward
        z1 = np.dot(X, W1) + b1
        a1 = sigmoid(z1)
        z2 = np.dot(a1, W2) + b2
        a2 = sigmoid(z2)

        # Error
        error = y - a2

        # Backprop
        d_a2 = error * sigmoid_derivative(a2)
        d_W2 = np.dot(a1.T, d_a2)
        d_b2 = np.sum(d_a2, axis=0, keepdims=True)

        d_a1 = np.dot(d_a2, W2.T) * sigmoid_derivative(a1)
        d_W1 = np.dot(X.T, d_a1)
        d_b1 = np.sum(d_a1, axis=0, keepdims=True)

        # Update bobot
        W1 += lr * d_W1
        b1 += lr * d_b1
        W2 += lr * d_W2
        b2 += lr * d_b2

    # Hitung loss akhir
    final_loss = np.mean(np.square(error))
    return final_loss

# -------------------------------------------
# Bandingkan konfigurasi awal vs baru
# -------------------------------------------
loss_hidden2 = train_model(hidden_size=2)
loss_hidden3 = train_model(hidden_size=3)

print("=== Perbandingan Loss ===")
print(f"Loss (hidden=2 neuron): {loss_hidden2}")
print(f"Loss (hidden=3 neuron): {loss_hidden3}")

if loss_hidden3 < loss_hidden2:
    print("Hidden layer 3 neuron memberikan hasil yang lebih baik!")
else:
    print("Hidden layer 2 neuron lebih stabil dalam percobaan ini.")
```

* Tambahkan fungsi aktivasi ReLU dan bandingkan hasil.

```python
import numpy as np

# Dataset XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# --- Aktivasi ---
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

# ---------------------------------------------------
# Fungsi training (bisa pilih 'sigmoid' atau 'relu')
# ---------------------------------------------------
def train_model(hidden_size, activation="sigmoid", lr=0.1, epoch_total=10000):

    input_size = 2
    output_size = 1

    # Inisialisasi bobot
    W1 = np.random.randn(input_size, hidden_size)
    b1 = np.zeros((1, hidden_size))
    W2 = np.random.randn(hidden_size, output_size)
    b2 = np.zeros((1, output_size))

    for epoch in range(epoch_total):

        # Forward
        z1 = np.dot(X, W1) + b1

        # Aktivasi hidden layer
        if activation == "sigmoid":
            a1 = sigmoid(z1)
        elif activation == "relu":
            a1 = relu(z1)

        z2 = np.dot(a1, W2) + b2
        a2 = sigmoid(z2)  # Output tetap sigmoid

        # Error
        error = y - a2

        # Backprop output layer
        d_a2 = error * sigmoid_derivative(a2)
        d_W2 = np.dot(a1.T, d_a2)
        d_b2 = np.sum(d_a2, axis=0, keepdims=True)

        # Backprop hidden layer pakai aktivasi yang dipilih
        if activation == "sigmoid":
            d_a1 = np.dot(d_a2, W2.T) * sigmoid_derivative(a1)
        elif activation == "relu":
            d_a1 = np.dot(d_a2, W2.T) * relu_derivative(z1)

        d_W1 = np.dot(X.T, d_a1)
        d_b1 = np.sum(d_a1, axis=0, keepdims=True)

        # Update bobot
        W1 += lr * d_W1
        b1 += lr * d_b1
        W2 += lr * d_W2
        b2 += lr * d_b2

    # Hitung loss akhir
    final_loss = np.mean(np.square(error))
    return final_loss

# ---------------------------------------------------
# Bandingkan Sigmoid vs ReLU
# ---------------------------------------------------
loss_sigmoid = train_model(hidden_size=3, activation="sigmoid")
loss_relu    = train_model(hidden_size=3, activation="relu")

print("=== Perbandingan Loss (Hidden = 3 neuron) ===")
print(f"Loss dengan Sigmoid: {loss_sigmoid}")
print(f"Loss dengan ReLU   : {loss_relu}")

if loss_relu < loss_sigmoid:
    print("ReLU lebih baik pada konfigurasi ini!")
else:
    print("Sigmoid lebih stabil dalam percobaan ini.")
```
